```python
#step 1 import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# step 2 read dataset
df=pd.read_csv("https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Boston.csv")
```

```python
# step 3 define output (y) and input(x)
df.columns
```

```python
y = df['MEDV']
X = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
```

```python
df.shape
```

```python
df.shape,X.shape,y.shape
```

```python
# step 4 train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
X_train
```

```python
# step 5 select model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
# step 6 train model --> able to finding best fit line
model.fit(X_train,y_train)
```

```python
model.intercept_ # y=c+mx in that c=31.25
```

```python
model.coef_  # give the slop of model y=c+mx in that m=
```

```python
# step 7 predict
y_pred=model.predict(X_test)
```

```python
# step 8 Evaluate
from sklearn.metrics import mean_absolute_percentage_error,mean_squared_error,mean_absolute_error
mean_absolute_percentage_error(y_test,y_pred) # model has 17% error
```

```python
mean_absolute_error(y_test,y_pred)
```
