```python
# step 1 import library
import pandas as pd
import matplotlib.pyplot as plt #(for data visualisation library)
```

```python
# Step 2 read dataset
df=pd.read_csv("https://github.com/YBIFoundation/Dataset/raw/main/Diabetes.csv")
```

```python
# step 3 define y and X
df.columns
```

```python
y=df['diabetes']
X=df[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
# step 4 train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
```

```python
# step 5 select model
from sklearn.tree import DecisionTreeClassifier,plot_tree
model=DecisionTreeClassifier(max_depth=2)
```

```python
# step 6 fit model
model.fit(X_train,y_train)
```

```python
#X_columns = ['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi', 'dpf', 'age']
plot_tree(model,feature_names=X.columns);
```

```python
# step 7 prediction
y_pred=model.predict(X_test)
```

```python
# step 8 evaluate
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
